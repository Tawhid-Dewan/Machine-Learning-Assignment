```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data
weight = np.array([2, 4, 5, 3, 6, 5, 7])
price = np.array([35, 60, 20, 50, 50, 55, 60])

# Create DataFrame
df = pd.DataFrame({"Weight": weight, "Price": price})

# --- Task 1: Manual Calculation of slope (m) and intercept (c) ---
x_mean = weight.mean()
y_mean = price.mean()

m = np.sum((weight - x_mean) * (price - y_mean)) / np.sum((weight - x_mean) ** 2)
c = y_mean - m * x_mean

# Predict price for weight = 6.5
pred_6_5 = m * 6.5 + c

# --- Task 2: Residuals ---
df["Prediction"] = m * df["Weight"] + c
df["Residual"] = df["Price"] - df["Prediction"]

# --- Task 3: MSE and MAE ---
MSE = np.mean(df["Residual"] ** 2)
MAE = np.mean(np.abs(df["Residual"]))

# --- Save to CSV ---
df.to_csv("Assignment_2_Data_sheet.csv", index=False)

# --- Plot ---
plt.figure(figsize=(7,5))
plt.scatter(df["Weight"], df["Price"], color="blue", label="Actual")
plt.plot(df["Weight"], df["Prediction"], color="red", label=f"Best Fit Line: Y={m:.2f}X+{c:.2f}")
plt.scatter(6.5, pred_6_5, color="black", marker="*", s=100, label=f"Prediction for 6.5kg = {pred_6_5:.2f}")
plt.xlabel("Weight")
plt.ylabel("Price")
plt.title("Price vs Weight")
plt.legend()
plt.grid(True)
plt.show()

(m, c, pred_6_5, MSE, MAE, df)
```
